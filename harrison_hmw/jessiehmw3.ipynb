## Problem 1: Show that an error of 3% in flux units is very nearly the same as an error of 0.03 magnitudes

Sounds good! We first know the equation to convert fluxes into magnitudes:

$$\frac{F_1}{F_2}=10^{0.4(m_2-m_1)}$$

We can then subtract another version of this where $F_1$ and $F_2$ are the same (as well as magnitudes). Note that we are converting from a percent/relative error ($\frac{|x_{\text{observed}}-x_{\text{true}}|}{x_{\text{true}}}$) to an absolute error ($|x_{\text{observed}}-x_{\text{true}}|$):

$$\frac{F_1-F_2}{F_2}=10^{0.4(m_2-m_1)}-10^{0.4(m_2-m_2)}\to F_{\text{err}}=10^{0.4(m_2-m_1)}-1$$
$$\to \log(F_{\text{err}}+1)=0.4(m_2-m_1) \to 2.5 \log(F_{\text{err}}+1)=m_{\text{err}}$$

Plugging in our value, we get:

$$2.5 \log(0.03+1)=~0.032$$

Fantastic!

## Problem 4: If a point source has a total count rate of 0.2 ph s−1, the sky background has a count rate of 0.5 ph s−1 pixel−1, the dark current is 10 electrons hour−1 pixel−1, and the read noise is 5 electrons, how many 1-minute exposures does it take to reach a S/N of 100? Assume the stellar PSF is distributed over 4 pixels.

Okay! This seems like a pretty good example of using the SNR function for detection, displayed here:

$$\frac{S}{N}=\frac{R_s t}{\sqrt{R_s t + n_{\text{pix}}(R_B t + R_D t + N_R^2)}}$$

From the question, we know that:

$$\frac{S}{N}=100$$
$$N_R^2=5\text{ e}^{-}\text{ pix}^{-1}$$
$$R_S=0.2\text{ }\gamma \text{ s}^{−1}$$
$$R_B=0.5\text{ }\gamma \text{ s}^{−1} \text{ pixel}^{−1}$$
$$R_D=10\text{ e}^{-}\text{ hr}^{-1}\text{ pix}^{-1}*\frac{1\text{ hr}}{3600\text{ s}} =2.78*10^{-3}\text{ e}^{-}\text{ s}^{-1}\text{ pix}^{-1}$$
$$n_{\text{pix}}=4$$

Let's plug in the numbers!!

$$100=\frac{(0.2\text{ }\gamma \text{ s}^{−1})t}{\sqrt{(0.2\text{ }\gamma \text{ s}^{−1})t + (4)((0.5\text{ }\gamma \text{ s}^{−1} \text{ pixel}^{−1}) t + (2.78*10^{-3}\text{ e}^{-}\text{ s}^{-1}\text{ pix}^{-1}) t + (5\text{ e}^{-}\text{ pix}^{-1})))}}\to 100 = \frac{0.2t}{\sqrt{0.2t+4(0.5t+5)}}$$

This requires some quadratic formula magic! Here's me solving it:

$$100 = \frac{0.2t}{\sqrt{0.2t+4(0.5t+5)}}\to 10^4=\frac{0.04t^2}{0.2t+4(0.5t+5)}\to 0.04t^2+10^4(0.2t+4(0.5t+5))=0$$
$$\to 0.04t^2+10^4(0.2t+2.0t)+5*10^{4}=0\to 0.04t^2+10^4(2.2)t+5*10^{4}=0$$
$$\to t=\frac{-b\pm\sqrt{b^2-4ac}}{2a}=\frac{-(2.2*10^4)\pm\sqrt{(2.2*10^4)^2-4(0.04)(5*10^4)}}{2(0.04)}\approx 550000\text{ s}$$

We get about 550000 seconds, or around 9166 one minute exposures (152 hours total)!
